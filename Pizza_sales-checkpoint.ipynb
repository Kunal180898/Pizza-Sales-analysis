```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("pizzas.csv")
df1.head()
```

```python
df2=pd.read_csv("pizza_types.csv",encoding='windows-1252')
df2.head()
```

```python
df3=pd.read_csv("orders.csv")
df3.head()
```

```python
df4=pd.read_csv("order_details.csv")
df4.head()
```

# 1. Sales Analysis

# 1.A.Total Sales

```python
merged_df = pd.merge(df4,df1, on='pizza_id')
merged_df.head()
```

```python
merged_df.shape
```

```python
merged_df["Revenue"] = merged_df["quantity"] * merged_df["price"]
merged_df.head(100)
```

```python
total_sales=merged_df["Revenue"].sum().round(2)
total_revenue_per_order = merged_df.groupby('order_id')['Revenue'].sum().reset_index()
# Calculate overall revenue
overall_revenue = total_revenue_per_order['Revenue'].sum()
# Display results
print("Total Revenue per Order:")
print(total_revenue_per_order)
print(f"Overall Revenue: ${overall_revenue:.2f}")
```

# 1.B.Sales by Pizza Type:

```python
merge_sales_pt=pd.merge(merged_df,df2,on="pizza_type_id")
merge_sales_pt.head()
```

```python
pt_sales=merge_sales_pt.groupby("name")["Revenue"].sum()
pt_sales=pt_sales.sort_values(ascending=True)
pt_sales
```

```python
plt.barh(pt_sales.index, pt_sales.values, height=0.5, color="#BB3E00")
plt.xlabel("Revenue")
plt.ylabel("Pizza type Name")
plt.title("Revenue as per Pizza Type")
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

# Conclusion :- The Thai Chicken Pizza generate most revenue        

# 1.C.Sales by Size

```python
merge_sales_pt.head()
```

```python
size_sales=merge_sales_pt.groupby("size")["Revenue"].sum()
size_sales
```

```python
plt.bar(size_sales.index, size_sales.values, width=0.5, color="#BB3E00")
plt.xlabel("Pizza Size")
plt.ylabel("Sales")
plt.title("Revenue as per Pizza Size")
plt.tight_layout()
plt.show()
```

# Conclusion:- L Size Pizzas generate more revenue than others.

# 1.D. Sales Trends Over Time:

```python
df3["date"]=pd.to_datetime(df3["date"])
df3["time"]=pd.to_timedelta(df3["time"])
df3.info()
```

```python
time_sales=pd.merge(merged_df,df3,on="order_id")
time_sales.head()
```

```python
time_sales['hour'] = time_sales['time'].dt.components['hours']
time_sales.head()
```

```python
hourly_sales=time_sales.groupby("hour")["Revenue"].sum()
hourly_sales
```

```python
plt.plot(hourly_sales.index,hourly_sales.values,marker=".",ms=20,mec="#BB3E00",mfc="#F7AD45")
plt.xlabel("Hours")
plt.ylabel("Sales")
plt.title("Sales Trend on hourly basis")
plt.show()
```

# Conclusion:- From 11 to 20 hrs have more sales

# 2. Order Patterns

```python
hourly_orders=time_sales.groupby("hour")["order_id"].count()
hourly_orders
```

```python
hourly_orders.plot(kind="bar",color="#BB3E00")
plt.title("Orders on the basis on hours")
plt.xlabel("Hours")
plt.ylabel("Orders")
plt.show()
```

# Conclusion:- At noon 12 to 13 and at evening 16 to 19 oders are high

# 5. Time-Based Analysis

# 5.A.Seasonal Trends

```python
time_sales.info()
```

```python
time_sales["month"]=time_sales["date"].dt.strftime('%B')
time_sales
```

```python
monthly_rev=time_sales.groupby("month")["Revenue"].sum().reset_index()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_rev['month'] = pd.Categorical(monthly_rev['month'], categories=month_order, ordered=True)
monthly_rev = monthly_rev.sort_values('month')
monthly_rev = monthly_rev.to_dict()
monthly_rev
```

```python
plt.plot(monthly_rev["month"].values(),monthly_rev["Revenue"].values(),marker=".",ms=20,mec="#BB3E00",mfc="#F7AD45")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.title("Trend of revenue on monthly basis")
plt.xticks(rotation=90)
plt.show()
```

# Conclusion:- From month March to August have more Sales

# 5.B.Order Frequency by Date:

# Conclusion: Most of Days has volume between 100 to 150

# 7.B.Heatmaps

    # for order frequency

```python
time_sales.info()
```

```python
time_sales2=time_sales
time_sales2["hour"]=time_sales2["time"].dt.components['hours']
time_sales2.head()
```

```python
od_freq_ht = time_sales2.groupby(["date","hour"])["order_id"].count().reset_index()
order_pivot = od_freq_ht.pivot(index='date', columns='hour', values='order_id').fillna(0)
order_pivot
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(order_pivot, cmap='YlOrRd', linewidths=0.5, linecolor='white')
plt.title('Sales Heatmap by Hour and Date')
plt.xlabel('Hour of the Day')
plt.ylabel('Date')
plt.show()
```

    # for sales

```python
sales_hr= time_sales2.groupby(["date","hour"])["Revenue"].sum().reset_index()
sales_pivot = sales_hr.pivot(index='date', columns='hour', values='Revenue').fillna(0)
sales_pivot
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(sales_pivot, cmap='YlOrRd', linewidths=0.5, linecolor='white')
plt.title('Sales Heatmap by Hour and Date')
plt.xlabel('Hour of the Day')
plt.ylabel('Date')
plt.show()
```
